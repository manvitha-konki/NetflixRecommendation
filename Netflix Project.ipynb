```python
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate
from surprise.model_selection import train_test_split
```

```python
# Prepare the dataset for the surprise library
reader = Reader(rating_scale=(1, 5))  # Specify rating scale
data = Dataset.load_from_df(final_df[['Cust_ID', 'Movie_ID', 'Rating']], reader)
```

```python
# Train-test split using surprise's method
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
# Train an SVD (Collaborative Filtering) model
model = SVD()
model.fit(trainset)
```

```python
# Function to get best-suited movie for a user in each genre
def get_best_movies_for_user(user_id, df, model):
    genres = df['Genre'].unique()  # Get unique genres
    best_movies = {}

    for genre in genres:
      # Filter movies by genre
      genre_movies_df = df[df['Genre'] == genre].reset_index(drop=True)
      genre_movies = genre_movies_df['Movie_ID'].unique()

      predictions = []
      for movie_id in genre_movies:
          pred = model.predict(user_id, movie_id, verbose=False)
          predictions.append((movie_id, pred.est))  # (Movie ID, Predicted Rating)

      # Sort by predicted rating and get the best one
      if predictions:
          best_movie = max(predictions, key=lambda x: x[1])
          movie_name = genre_movies_df[genre_movies_df['Movie_ID'] == best_movie[0]]['Name'].iloc[0]
          best_movies[genre] = (movie_name, best_movie[1])  # (Movie Name, Predicted Rating)

    return best_movies

# Example: Get the best movies for a specific user
user_id = 2632461
best_movies = get_best_movies_for_user(user_id, final_df, model)

# Display results
print("Best Movies for User in Each Genre:")
for genre, (movie_name, rating) in best_movies.items():
    print(f"Genre: {genre} ==> Movie: {movie_name}, Predicted Rating: {rating:.2f}")
```

```python
# Group by Genre and calculate average ratings
genre_ratings = final_df.groupby('Genre')['Rating'].mean()

# Find the best and worst genres
best_genre = genre_ratings.idxmax()
worst_genre = genre_ratings.idxmin()

# Display results
print("Average Ratings by Genre:")
print(genre_ratings)

print(f"\nBest Genre: {best_genre} with an average rating of {genre_ratings[best_genre]}")
print(f"Worst Genre: {worst_genre} with an average rating of {genre_ratings[worst_genre]}")
```
